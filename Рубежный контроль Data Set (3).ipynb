```python
# Аггрегированные признаки по артистам
artist_stats = df.groupby('artists').agg(
    success_share=('is_successful', 'mean'),
    avg_energy=('energy', 'mean'),
    avg_danceability=('danceability', 'mean'),
    avg_valence=('valence', 'mean'),
    avg_duration_min=('duration_min', 'mean')
).reset_index()

# Делим артистов на 2 группы: выше/ниже средней доли хитов
mean_share = artist_stats['success_share'].mean()
high = artist_stats[artist_stats['success_share'] > mean_share]
low  = artist_stats[artist_stats['success_share'] <= mean_share]

feat_cols = ['avg_energy','avg_danceability','avg_valence','avg_duration_min']

high_means = high[feat_cols].mean()
low_means  = low[feat_cols].mean()

profile_df = pd.DataFrame({
    'Признак': ['Energy','Danceability','Valence','Длительность (мин)'],
    'Артисты с высокой долей хитов': high_means.values,
    'Артисты с низкой долей хитов': low_means.values
})

profile_df.set_index('Признак').plot(kind='bar', figsize=(8,5), color=['steelblue','lightgray'])
plt.title("Аудио-профиль артистов с разной долей хитов")
plt.ylabel("Среднее значение признака")
plt.xticks(rotation=0)
plt.show()
```

значения около 0 означают, что линейной связи почти нет.

Бизнес-вопрос:
Какие особенности аудио‑профиля отличают успешных артистов (с долей популярных треков выше средней) и как эти инсайты могут помочь A&R-менеджерам в поиске талантов?

1. Ключевые наблюдения (Data Insights)
КАнализ показывает сильный дисбаланс: подавляющее большинство артистов имеют малый каталог (1–5 треков). При этом «хиты» (топ-20% по популярности) чаще концентрируются у небольшой группы опытных исполнителей с обширным каталогом (21+ треков).

Фактор длительности:
Единственное заметное отличие в аудио-профиле — это хронометраж. У артистов с высокой долей успешных треков композиции в среднем немного короче («компактнее»), чем у менее успешных.
Вывод: Лаконичность коррелирует с успехом в стриминге.

Жанровый контекст:
Хиты распределены по жанрам достаточно равномерно, но выделяются лидеры: pop-film, k-pop, chill, hip-hop*. Ни один жанр не забирает 100% успеха, но попадание в трендовый жанр повышает шансы.

2. Рекомендации для A&R (Стратегия отбора)

Опираясь на графики, лейблам не стоит искать артистов, ориентируясь только на «звуковые параметры» (например, искать только самые энергичные треки). Вместо этого эффективная A&R-стратегия должна быть комбинированной:

1. Приоритет артистам, которые способны генерировать контент регулярно (размер каталога > 5 треков), так как хиты кучкуются у продуктивных авторов.
2. Отдавать предпочтение трекам с умеренной длительностью (избегать затянутых композиций), так как они лучше перформят в стриминге.
3. Использовать жанровые тренды (K-Pop, Pop, Hip-Hop) как фильтр первого уровня, но не как гарантию успеха.
4. Не отбрасывать артистов только потому, что их треки «слишком грустные» (valence) или «недостаточно танцевальные», так как данные подтверждают, что хиты возможны с любым настроением.

Финальный вывод:
Успех в Spotify не поиск «идеальной волны» на графике, а работа с историей артиста . Данные доказывают, что «магического аудио-параметра» не существует, но существует паттерн продуктивности и формата.
